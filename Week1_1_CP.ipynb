```python
t = int(input().strip())

for i in range(t):
    n = int(input().strip())
    meetings = []

    for j in range(n):
        s, e = map(int, input().split())
        meetings.append((s, e))

    # Sort meetings by end time
    meetings.sort(key=lambda x: x[1])

    count = 1
    last_end = meetings[0][1]

    for i in range(1, n):
        if meetings[i][0] >= last_end:
            count += 1
            last_end = meetings[i][1]

    print(count)
```
